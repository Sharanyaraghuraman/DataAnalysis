```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10,5))
plt.boxplot(df["Police_Response_Time"],showmeans=True,vert=True)
plt.title("boxplot")
plt.xlabel("Police Response Time")
plt.savefig("Sharanya_IDAproject_boxplot")
plt.show()
```

```python
df2=df.iloc[:500,::]
df2
```

```python
plt.bar(df2["State"],df2["Police_Response_Time"],label="Police response time for different states",color=["black"])
plt.title("barplot")
plt.xlabel("State")
plt.ylabel("Police Response Time")
plt.xticks(rotation=90, fontsize=6)
plt.yticks(fontsize=7)
plt.savefig("Sharanya_ida_barplot")
plt.legend()
plt.show()
```
